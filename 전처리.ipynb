question에 nan이 존재 -> 질문을 title에 적었기 때문

```python
df.columns = ['i', 'title', 'question', 'answer', 'url', 'index']
df.drop(['i', 'index'], axis=1, inplace=True)
```

```python
# index 수정
df.reset_index(inplace=True)
df
```

### title과 question이 중복인 경우 제목을 삭제
- title과 question을 합쳐서 하나의 question으로 만들 것이기 때문

```python
# title과 question이 중복인 경우 제목 삭제 -> question이 null이 아닌 경우만
b = list(df[df.iloc[:,2].isnull()]['index'])
```

```python
# title과 question이 중복인 경우 제목 삭제
for i in range(0, len(df)):
  if (i not in b):
    if (df.iloc[i,1][:] == df.iloc[i,2][0:len(df.iloc[i,1])]):
      df.iloc[i,1] = np.nan
```

```python
df.isnull().sum()  # 1058개의 중복 제목 제거
```

### title + question -> question

```python
# 제목과 질문을 합쳐서 question' 하나의 열로 만들기
df['Q'] = df['title'].str.cat(df['question'],sep=" ", na_rep="")
df = df.drop(['title', 'question'], axis=1)
df.rename(columns = {'Q': 'question'}, inplace=True)
df = df[['index', 'question', 'answer', 'url']]
df
```

```python
print(df.isnull().sum())
print(df.shape)
```

### '\n"을 "\0"(blank)로 바꾸기

```python
# '\n"을 "\0"(blank)로 바꾸기
df['question'] = df['question'].str.replace('\r\n', ' ')
df['question'] = df['question'].str.replace('\n', ' ')
df['question']
```

```python
df['answer'] = df['answer'].str.replace('\r\n', ' ')
df['answer'] = df['answer'].str.replace('\n', ' ')
df['answer']
```

### 수의사 답변 2개 이상 처리

```python
# 수의사 답변 2개 이상인 부분
df[df['answer'].str.contains('@Excepiton')]
```

```python
# 중복 답변 길이
len(df[df['answer'].str.contains('@Excepiton')]['question'])
```

1개의 질문과 2개의 답변 -> 2개의 질문&답변으로 분리
* 1) 질답 복제 -> 인덱스 수정
* 2) 맨 처음 '####' 찾아서 삭제 ('Exception 수의사 답변이 2개 이상')
* 2) 앞부분 답변: '####' 찾아서 앞부분만 저장
* 3) 뒷부분 답변: '####' 찾아서 뒷부분만 저장

```python
# 질답 복제 -> 인덱스 수정
df = df.append(df[df['answer'].str.contains('@Excepiton')], ignore_index=True)
df.reset_index(inplace=True)
df.drop(['index'], axis=1, inplace=True)
df.rename(columns = {'level_0': 'index'}, inplace=True)
df
```

```python
# 22921개의 행 -> 수의사 답변 2개인 40개의 행을 복제해서 총 22961개의 행이 됨
```

```python
# '####' 찾아서 맨 처음 '입니다' 삭제 ('Exception 수의사 답변이 2개 이상입니다')
c = list(df[df['answer'].str.contains('@Excepiton')]['index'])
for i in range(0, len(df)):
  if (i in c):
    point = df['answer'][i].find('####')
    df['answer'][i] = df['answer'][i][point+5:]
```

```python
# 다시 '####' 찾아서 앞부분 답변만 저장
for i in range(0, len(df)- 40):
  if (i in c):
    point = df['answer'][i].find('####')
    df['answer'][i] = df['answer'][i][:point]
```

```python
# 다시 '####' 찾아서 뒷부분 답변만 저장
for i in range(len(df)- 40, len(df)):
  if (i in c):
    point = df['answer'][i].find('####')
    df['answer'][i] = df['answer'][i][point+5:]
```
