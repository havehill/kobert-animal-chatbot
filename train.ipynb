```python
# ★ Hugging Face를 통한 모델 및 토크나이저 Import
from kobert_tokenizer import KoBERTTokenizer
from transformers import BertModel

from transformers import AdamW
from transformers.optimization import get_cosine_schedule_with_warmup
import pytorch_lightning as pl
from sklearn.metrics import accuracy_score
from torch.utils.data import Dataset
from torch.utils.data.dataset import random_split

device = torch.device("cuda:0")

tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
tokenizer.encode('한국어 모델을 공유합니다.')
```

```python
text = '한국어 모델을 공유합니다.'
model = BertModel.from_pretrained('skt/kobert-base-v1')

inputs = tokenizer.batch_encode_plus([text])
out = model(input_ids = torch.tensor(inputs['input_ids']), attention_mask = torch.tensor(inputs['attention_mask']))
```

```python
from kobert_transformers import get_tokenizer

file = open('/content/drive/MyDrive/자진프/kobert-wellness-chatbot/data/dialog_text_for_classification.txt', 'r', encoding='utf-8')
tokenizer = tokenizer if tokenizer is not None else get_tokenizer()
num = 0
total_num = -1
idx_list = []


while True:
    line = file.readline()
    if not line:
        break
    datas = line.split("\t")

    # Label
    label = int(datas[1][:-1])
    total_num += 1

    index_of_words = tokenizer.encode(datas[0])
    if len(index_of_words) > 512:
        num += 1
        idx_list.append(total_num)
        print(len(index_of_words))
        print(datas[0])

print("\n512 넘는 토큰 수 : ",num)
print("\n512 넘는 토큰 인덱스 : ", idx_list)
print("\n총개수 : ", total_num+1)
```
